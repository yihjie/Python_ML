f(x,y) = 1 - 0.5x^2 - 0.5y^2

0 <= x < 1
0 <= y < 1

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D                       #  3D 
import numpy as np

def f(x, y):                                                  # 
    return (1 - (x ** 2) / 2 - (y ** 2) / 2)

ax  = plt.axes(projection='3d')

X = np.arange(0, 1, 0.01)                                     #  X 
Y = np.arange(0, 1, 0.01)                                     #  Y 
X, Y  = np.meshgrid(X, Y)                                     # 
ax.plot_surface(X, Y, f(X, Y), color='lightgreen')            #  3D 

plt.axis('equal')
plt.grid()
plt.show()
```

f(x,y) = x^2 + y^2

-3 <= x < 3
-3 <= y < 3

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def f(x, y):
    return (np.power(x, 2) + np.power(y, 2))

ax  = plt.axes(projection='3d')

X = np.arange(-3, 3, 0.1)
Y = np.arange(-3, 3, 0.1)
X, Y = np.meshgrid(X, Y)

ax.plot_surface(X, Y, f(X, Y), cmap='hsv')
ax.set_xlabel('x', color='b')
ax.set_ylabel('y', color='b')
ax.set_zlabel('z', color='b')

plt.grid()
plt.show()
```

f(x, y) = sin((x^2 + y^2)^0.5)

-3 <= x < 3
-3 <= y < 3

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def f(x, y):
    r = np.sqrt(np.power(x, 2) + np.power(y, 2))
    return (np.sin(r))

ax = plt.axes(projection='3d')


X = np.arange(-3, 3, 0.1)
Y = np.arange(-3, 3, 0.1)
X, Y = np.meshgrid(X, Y)

ax.plot_surface(X, Y, f(X, Y), cmap='rainbow')
ax.set_xlabel('x', color='b')
ax.set_ylabel('y', color='b')
ax.set_zlabel('z', color='b')

plt.grid()
plt.show()
```
