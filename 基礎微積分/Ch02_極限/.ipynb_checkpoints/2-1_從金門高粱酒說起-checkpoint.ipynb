稀釋金門高粱酒的酒精濃度

```python
alchol = 58

for x in range(0, 11):
    if x > 0 :
        alchol *= 0.5
    print(f"當 x = {x:2d}, 酒精濃度 = {alchol}")
```

```python
import matplotlib.pyplot as plt

alchol = 58

x = [x for x in range(0, 11)]          # 調淡酒精的次數
y = [alchol * (1/2) ** y for y in x]   # 酒精濃度

plt.axis([0, 12, 0, 60])
plt.plot(x, y)
plt.xlabel("Times")
plt.ylabel("Alcohol concentration")

plt.grid()
plt.show()
```
