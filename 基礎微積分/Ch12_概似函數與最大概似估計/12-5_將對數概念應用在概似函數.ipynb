L(p) vs ln L(p)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.01, 1, 1000)
y = np.log(x)                      # e 為底數的 log()

plt.plot(x, y, color='b')

plt.xlabel('L(p)')
plt.ylabel('ln L(p)', rotation=0)

plt.xlim(left=0)
plt.xlim(right=1)
plt.ylim(top=0)
plt.grid()
plt.show()
```
