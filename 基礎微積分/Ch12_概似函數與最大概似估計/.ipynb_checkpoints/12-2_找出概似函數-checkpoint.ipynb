L(p) = p * p

0 <= p < 1
x = p
y = L(p)


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 1000)
y = x * x

plt.plot(x, y, color='b')
plt.xlabel('Probability')
plt.ylabel('Likelihood')


plt.xlim(left=0)
plt.xlim(right=1)
plt.ylim(bottom=0)
plt.ylim(top=1)
plt.grid()
plt.show()
```

L(p) = p * p * ( 1 - p )

0 <= p < 1
x = p
y = L(p)

p 的最大值 
f(x) = -x^3 + x^2
f'(x) = -3x^2 + 2x = x(-3x+2)
x = 0 ; x = 2/3


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 1000)
y = x * x * (1 - x)

# 最大值
x_max = 2 / 3
y_max = x_max * x_max * ( 1 - x_max)

# 切線
x_new = [x_max - 0.05, x_max, x_max + 0.05]
y_new = [ y_max for x in x_new]
plt.plot(x_new, y_new, color='r')

plt.plot(x_max, y_max, '-o', color='r')
plt.text(x_max - 0.1 , y_max + 0.01, '(' + str(np.round(x_max,4)) + ', ' + str("{:.2%}".format(y_max)) + ')')

plt.plot(x, y, color='b')
plt.xlabel('Probability')
plt.ylabel('Likelihood')


plt.xlim(left=0)
plt.xlim(right=1)
plt.ylim(bottom=0)

plt.grid()
plt.show()
```
