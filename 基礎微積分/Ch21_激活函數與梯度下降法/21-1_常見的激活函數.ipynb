Sigmoid Function ( S-Func )
    f(x) = 1 / (1 + exp(-x))

Tanh Function
    f(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x))
    
ReLu Function
    f(x) = 0 for x < 0
    f(x) = x for x >= 0

```python
# Tanh Function
import matplotlib.pyplot as plt
import numpy as np

def tanh(x):
    return ( (np.e ** x - np.e ** (-x)) / (np.e ** x + np.e ** (-x)) )

x = np.linspace(-5, 5, 10000)
y = [tanh(x) for x in x]

plt.axis([-5, 5, -1, 1])
plt.plot(x, y, label='Tanh function')

plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
# ReLu Function
import matplotlib.pyplot as plt
import numpy as np

def relu(x):
    return ( x if x >= 0 else 0 )

x = np.linspace(-5, 5, 10000)
y = [relu(x) for x in x]

plt.axis([-5, 5, -5,5])
plt.plot(x, y, label='ReLu function')

plt.legend(loc='best')
plt.grid()
plt.show()
```
