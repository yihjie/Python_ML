```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 10000)
y = 1 / ((2 * np.pi) ** 0.5 ) * np.exp(-x ** 2 / 2)

plt.plot(x, y, color='b')

plt.fill_between(x, y1=y, y2=0, where= (x >= -0.8) & ( x < -0.6), facecolor='green')

plt.xlim(left=-3)
plt.xlim(right=3)
plt.ylim(bottom=0)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
from scipy.integrate import trapz

x = np.linspace(-3, 3, 10000)
y = st.norm.pdf(x)

plt.plot(x, y)

xs = np.linspace(-0.8, -0.6, 10000)
p  = trapz(st.norm.pdf(xs), xs)

print(f"落在 -0.8 與 -0.6 之間的機率是 {100 * p : 4.2f} %")
plt.fill_between(xs, st.norm.pdf(xs), color='green')

plt.xlim(left=-3)
plt.xlim(right=3)
plt.ylim(bottom=0)
plt.grid()
plt.show()
```
