```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 10000)
y = 1 / ((2 * np.pi) ** 0.5 ) * np.exp(-x ** 2 / 2)

plt.plot(x, y, color='b')


plt.xlim(left=-3)
plt.xlim(right=3)
plt.ylim(bottom=0)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st

x = np.linspace(-3, 3, 10000)

plt.plot(x, st.norm.pdf(x, loc=0, scale=1), color='b')


plt.xlim(left=-3)
plt.xlim(right=3)
plt.ylim(bottom=0)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st

x = np.linspace(-3, 3, 10000)

loc_scale_dict = {0:1, -1:2, 1:0.5}

for loc, scale in loc_scale_dict.items():
    plt.plot(x, st.norm.pdf(x, loc=loc, scale=scale))


plt.xlim(left=-3)
plt.xlim(right=3)
plt.ylim(bottom=0)
plt.grid()
plt.show()
```
