y = f(x) = 3x^2 - 12x + 10
求 y 的極小值
求 斜率

```python
import matplotlib.pyplot as plt
import numpy as np

a = 3
b = -12
c = 10

# 計算微分
x_min = -b / ( 2 * a)

# 將 x_min 帶入二次函數
y_min = a * x_min ** 2 + b * x_min + c

plt.text( x_min - 0.25, y_min - 1.2, '(' + str(x_min) + ',' + str(y_min) + ')' )
plt.plot( x_min, y_min, '-o', color='r')


# 繪製此函數圖形
x = np.linspace(0, 4, 50)
y = a * x ** 2 + b * x + c

plt.plot(x, y, color='b')

# 導數 ( derivative )
a_de = 2 * a
b_de = b
x_de = np.linspace(0, 4, 50)
y_de = a_de * x + b_de

# 繪製導數
plt.plot(x_de, y_de, color='g')

# 導數為 0 的 (x,y)座標
x_zero = -b_de / a_de
y_zero = 0.0

plt.text(x_zero - 0.25, y_zero + 1.2, '(' + str(x_zero) + ', ' + str(y_zero) + ')')
plt.plot(x_zero, y_zero, '-o', color='r')

plt.grid()
plt.show()
```
