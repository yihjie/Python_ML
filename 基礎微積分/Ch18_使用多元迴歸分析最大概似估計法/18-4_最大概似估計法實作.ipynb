```python
import matplotlib.pyplot as plt
import numpy as np


design_days = { 1:85,                                   # 設計師預估裝潢天數
                2:91,
                3:76,
                4:102,
                5:68,
                6:72,
                7:66 }
ave = np.array(list(design_days.values())).mean()       # 平均天數

x = np.linspace(0, 200, 200)                            # 陣列 x
y = np.linspace(0, 200, 200)                            # 陣列 y
min_x = 100                                             # 預估最初預估極值的 x
min_y = 100                                             # 預估最初預估極值的 y

for i in range(len(x)):
    sum = 0
    for j in range(len(design_days)):
        sum += ( ave -x[i]) ** 2
    y[i] = sum
    if y[i] < min_y:
        min_x = i
        min_y = y[i]
        
plt.plot(x, y)
plt.plot(min_x, min_y, '-o')
plt.text(min_x - 20, min_y + 2000, '(' + str(min_x) + ', ' + str(round(min_y, 5)) + ')')

plt.xlim(left=0)
plt.xlim(right=200)
plt.ylim(bottom=0)
plt.grid()
plt.show()
```
