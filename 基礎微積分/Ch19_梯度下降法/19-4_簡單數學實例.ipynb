```python
import matplotlib.pyplot as plt
import numpy as np
import time

rate = eval(input("請輸入學習率 : "))
init_x = eval(input("請輸入參數值 : "))

x = np.arange(-10, 10, 0.1)
y = x ** 2

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'rate = {rate}     initial = {init_x}')

new_x = 0
old_x = 0
for i in range(1, 16):              # 15 圈
    if i == 1:
        old_x = init_x
    else:
        old_x = new_x
    new_x = old_x - rate * 2 * old_x # 計算新的參數
    
    plt.plot([old_x, new_x], [old_x ** 2, new_x ** 2], 'go-')
    print(f'loop = {i:2d}, old_x = {old_x:4.2f}, new_x = {new_x:4.2f}')
    time.sleep(1)
    
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import time

rate = eval(input("請輸入學習率 : "))
init_x = eval(input("請輸入參數值 : "))

def L_fun(x):
    return ( 7 * x ** 2 - 1120 * x + 45850 )

def LL_fun(x):
    return ( 14 * x - 1120)

x = np.arange(50, 100, 0.1)
y = L_fun(x)



plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'rate = {rate}     initial = {init_x}')

new_x = 0
old_x = 0
for i in range(1, 31):              # 30 圈
    if i == 1:
        old_x = init_x
    else:
        old_x = new_x
    new_x = old_x - rate * LL_fun(old_x) # 計算新的參數
    
    plt.plot([old_x, new_x], [L_fun(old_x) , L_fun(new_x)], 'go-')
    print(f'loop = {i:2d}, old_x = {old_x:4.2f}, new_x = {new_x:4.2f}')
    time.sleep(1)
    
plt.grid()
plt.show()
```
