```python
import matplotlib.pyplot as plt
import numpy as np

# 二次函數的係數
a = -3.5
b = 18.5
c = -5

# 標記業績點
xy_point = {1:10, 2:18, 3:19}
for x,y in xy_point.items():
    plt.text(x + 0.05, y - 1, '(' + str(x) + ', ' + str(y) + ')')
    plt.plot(x, y, '-o', color='g')

# 繪製此函數圖形
x = np.linspace(0, 4, 50)
y = a * x ** 2 + b * x + c
plt.plot(x, y, color='b')

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 二次函數的係數
a = -3.5
b = 18.5
c = -5

# 標記業績點
xy_point = {1:10, 2:18, 3:19}
for x,y in xy_point.items():
    plt.text(x + 0.05, y - 1, '(' + str(x) + ', ' + str(y) + ')')
    plt.plot(x, y, '-o', color='g')

# 繪製此函數圖形
x = np.linspace(0, 4, 50)
y = a * x ** 2 + b * x + c
plt.plot(x, y, color='b')

# 微分
x_max = round( (-b / ( 2 * a )), 2)                           # 行銷次數 x 軸值
y_max = round( (a * x_max ** 2 + b * x_max + c), 2)    # 業績增加最大值
plt.text(x_max -0.4, y_max - 1.5, '(' + str(x_max) + ', ' + str(y_max) + ')')
plt.plot(x_max, y_max, '-o', color='r')

plt.grid()
plt.show()
```
