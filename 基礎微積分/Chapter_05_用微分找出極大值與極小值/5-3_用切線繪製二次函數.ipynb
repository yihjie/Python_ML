5-3-3 繪製二次函數與切線

y = f(x) = 3x^2 - 12x + 10

```python
import matplotlib.pyplot as plt
import numpy as np

a = 3
b = -12
c = 10

# 繪製此函數圖形
x = np.linspace(0, 4, 50)
y = a * x ** 2 + b * x + c
plt.plot(x, y, color='b')

# 繪製經過 x = 0, 1, 2, 3, 4 的切線
for x_loc in range (0, 5):
    y_loc = a * x_loc ** 2 + b * x_loc + c                  # y 座標
    slope = 2 * a * x_loc + b                               # 每一點的斜率

# x_new 和 y_new 是經過切線的座標，只取 3 點連線
    x_new = [ x_loc - 1, x_loc, x_loc + 1]
    y_new = [ slope * (x - x_loc) + y_loc for x in x_new ]
    plt.plot(x_new, y_new, color='g')

plt.grid()
plt.show()
```

2x + 2y = 100
x > 0
y > 0
求 x * y 的最大值

```python
import matplotlib.pyplot as plt
import numpy as np

# 二次函數的係數
a = -1
b = 50

# 計算微分
x_max  = -b / ( 2 * a )

# 將 x_max 帶入二次函數
y_max = a * x_max ** 2 + b * x_max

plt.text(x_max - 5, y_max -50, '(' + str(x_max) + ', ' + str(y_max) + ')')
plt.plot(x_max, y_max, '-o', color='r')

# 繪製此函數圖形
x = np.linspace(0 ,51, 50)
y = a * x ** 2 + b * x
plt.plot(x, y, color='b')

plt.grid()
plt.show()
```
