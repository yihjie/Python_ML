F(X) = (X+1)dX
F(5) - F(1)

```python
import matplotlib.pyplot as plt
import numpy as np

# 原始函數 F(x) 的係數
a = 1
b = 1

# 繪製此函數積分區間圖形
x = np.linspace(0, 6, 1000)
y = a * x + b
plt.plot(x, y, color='b')
plt.fill_between(x, y1 = y, y2 = 0, where = (x >= 1) & (x <= 5), facecolor='lightgreen')

plt.xlim(left=0)
plt.ylim(bottom=0)
plt.grid()
plt.show()
```

f(3) - f(1)

```python
from sympy import *

x = Symbol('x')
f = x ** 2
print(integrate(f, (x, 1, 3)))
```

f(x) = (-x^2 + 2x)dx
f(4) - f(-2) = f(4) - f(2) + f(2) - f(0) + f(0) - f(-2)

```python
from sympy import *

x = Symbol('x')
f = -x ** 2 + 2 * x
f1 = x ** 2 - 2 * x

sec1 = integrate(f1, (x, -2, 0))
sec2 = integrate(f, (x, 0, 2))
sec3 = integrate(f1, (x, 2, 4))
print(sec1 + sec2 + sec3)
```
