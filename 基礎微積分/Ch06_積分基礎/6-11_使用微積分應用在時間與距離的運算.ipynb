磚塊飛行軌跡

y = f(x) = -3x^2 + 18x
f'(x) = -6x + 18
所以 x = 3 會有最高點 f(3) = -27 + 54 = 27

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

# 原始函數 f(x) 的係數
a = -3
b = 18

# 最高點座標
x_max = -b / ( 2 * a)
y_max = a * x_max ** 2 + b * x_max
plt.text(x_max - 0.5,  y_max - 2 , '(' + str(x_max) + ', ' + str(y_max) + ')')
plt.plot(x_max, y_max, '-o', color='r')

# 繪製此函數積分區間圖形
x = np.linspace(0, 6, 1000)
y = a * x ** 2 + b * x
plt.plot(x, y, color='b')

plt.grid()
plt.show()

x = Symbol('x')
f = a * x ** 2 + b * x
print("f'(x) = ", f.diff())
```

```python
最初飛行速度
f'(0) = 18 - 6 * 0 = 18
```
