## 解一元二次方程式的根

f(x) = 3x^2 + 5x + 1 = 0

```python
from scipy.optimize import root

def f(x):
    return ( a * x ** 2 + b * x + c )

a = 3
b = 5
c = 1

r1 = root(f, 0)                 # 初始迭代值 0
print(r1.x)
r2 = root(f, -1)                # 初始迭代值 -1
print(r2.x)
r3 = root(f, 1)                 # 初始迭代值 1
print(r3.x)
```

## 計算 2 個線性方程式的交叉點

f(x) = 1x^2 - 5x + 7
g(x) = 0x^2 + 2x + 1

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import root

def f(x):
    return ( a1 * x ** 2 + b1 * x + c1 )

def g(x):
    return ( a2 * x ** 2 + b2 * x + c2 )

# f(x) 係數
a1 = 1
b1 = -5
c1 = 7

# g(x) 係數
a2 = 0
b2 = 2
c2 = 1


# 計算交叉點
r1 = root(lambda x:f(x) - g(x), 0)                # 初始迭代值 0
r2 = root(lambda x:f(x) - g(x), 5)                # 初始迭代值 5

print("x1 = %4.2f, y1 = %4.2f" % ( r1.x, f(r1.x)))
print("x2 = %4.2f, y2 = %4.2f" % ( r2.x, g(r2.x)))

# 繪製 f(x) 圖形
x1 = np.linspace(0, 10, 100)
y1 = a1 * x1 ** 2 + b1 * x1 + c1
plt.text(r1.x - 0.5, f(r1.x) + 3, '(' + str(np.round(float(r1.x), 2)) + ', ' + str(np.round(float(f(r1.x)), 2)) + ')')
plt.plot(r1.x, f(r1.x), 'o')
plt.plot(x1, y1, '-', label= 'f(x) = ' + str(a1) + 'x^2 + ' + str(b1) + 'x + ' + str(c1))

# 繪製 g(x) 圖形
x2 = np.linspace(0, 10, 100)
y2 = a2 * x2 ** 2 + b2 * x2 + c2
plt.text(r2.x + 0.1, g(r2.x) - 3, '(' + str(np.round(float(r2.x), 2)) + ', ' + str(np.round(float(g(r2.x)), 2)) + ')')
plt.plot(r2.x, g(r2.x), 'o')
plt.plot(x2, y2, '-', label = 'g(x) = ' + str(a2) + 'x^2 + ' + str(b2) + 'x + ' + str(c2))

plt.legend(loc='best')
plt.xlim(left=0)
plt.xlim(right=10)
plt.ylim(bottom=0)
plt.show()
```
