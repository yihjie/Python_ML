f(x) = 0.5x^2 + x + C

```python
import matplotlib.pyplot as plt
import numpy as np

# 原始函數 F(x) 的係數
a = 0.5
b = 1

# 繪製此函數圖形
for C in range(1, 4):
    x = np.linspace(-4, 4, 100)
    y = a * x ** 2 + b * x + C
    plt.plot(x, y, color='b')

plt.grid()
plt.show()
```
