f(x) = x^2 - 2
g(x) = -x^2 + 2x + 2

```python
import matplotlib.pyplot as plt
import numpy as np


# 共同範圍
x = np.linspace(-2, 3, 1000)

# 原始函數 f(x) 的係數
a1 = 1
c1 = -2

y1 = a1 * x ** 2 + c1
plt.plot(x, y1, color='b')
plt.text(3 - 0.4, a1 * ( 3 ** 2) + c1 - 0.5, 'f(x)')

# 原始函數 g(x) 的係數
a2 = -1
b2 = 2
c2 = 2

y2 = a2 * x ** 2 + b2 * x + c2
plt.plot(x, y2, color='g')
plt.text(3 - 0.4, a2 * ( 3 ** 2) + b2 * 3 + c2 - 0.5, 'g(x)')

plt.fill_between(x, y1=y1, y2=y2, where= ( x >= -1 ) & ( x <= 2 ), facecolor='yellow')

plt.grid()
plt.show()
```

h(x) = g(x) - f(x)
求 S ( h(x)dx )
g(x) = -x^2 + 2x + 2
f(x) = x^2 -2
h(x) = g(x) - f(x) = -2x^2 + 2x + 4

```python
from sympy import *

x = Symbol('x')
h = -2 * x ** 2 + 2 * x + 4

print(integrate(h, (x, -1, 2)))
```
