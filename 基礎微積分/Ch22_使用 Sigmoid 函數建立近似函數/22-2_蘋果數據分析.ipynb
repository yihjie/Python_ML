```python
from sympy import solve, Symbol
import numpy as np

rate1 = 0.04
rate2 = 0.41

coef1 = 1
coef2 = 4.4

a1 = Symbol('a1')
b1 = Symbol('b1')

eq1 = np.log(rate1 / (1 - rate1)) - coef1 * a1 - b1
eq2 = np.log(rate2 / (1 - rate2)) - coef2 * a1 - b1
data = solve((eq1, eq2), (a1, b1))

print(f'a1 = {data[a1]:5.3f}')
print(f'b1 = {data[b1]:5.3f}')
```

```python
from sympy import solve, Symbol
import numpy as np

rate1 = 0.99
rate2 = 0.6666

coef1 = 4.4
coef2 = 5

a1 = Symbol('a1')
b1 = Symbol('b1')

eq1 = np.log(rate1 / (1 - rate1)) - coef1 * a1 - b1
eq2 = np.log(rate2 / (1 - rate2)) - coef2 * a1 - b1
data = solve((eq1, eq2), (a1, b1))

print(f'a1 = {data[a1]:5.3f}')
print(f'b1 = {data[b1]:5.3f}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1.0, 5.0, 21)
y = [4, 4, 5, 6, 8, 12, 18, 22, 35, 40, 45,
    33, 28, 30, 32, 36, 38, 41, 35, 31, 20]

plt.scatter(x,y)
plt.xlabel('Quality')
plt.ylabel('rate')

x1 = np.linspace(1.0, 5.0, 1000)
y1 = [x for i in x1]
y2 = [x for i in x1]
y3 = [x for i in x1]
for i in range(len(x1)):
    y1[i] = 1 / ( 1 + np.exp(-0.828 * x1[i] + 4.006)) * 100
    y2[i] = 1 / ( 1 + np.exp(6.504 * x1[i] - 33.212)) * 100
    y3[i] = y1[i] + y2[i]
plt.plot(x1, y1, color='g')
plt.plot(x1, y2, color='purple')
plt.plot(x1, y3, color='r')

plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1.0, 5.0, 1000)
y = [x for i in x]

for i in range(len(x)):
    f1 = 1 / ( 1 + np.exp(-0.828 * x[i] + 4.006))
    f2 = 1 / ( 1 + np.exp(6.504 * x[i] - 33.212))
    y[i] = f1 + f2 - 1
    
plt.plot(x, y)
plt.xlabel('Quality')
plt.ylabel('rate')
plt.grid()
plt.show()
```
