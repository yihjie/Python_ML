```python
# f3(x)

from sympy import solve, Symbol
import numpy as np

rate1 = 0.01
rate2 = 0.99

coef1 = 2.6
coef2 = 3.0

a1 = Symbol('a1')
b1 = Symbol('b1')

eq1 = np.log(rate1 / (1 - rate1)) - coef1 * a1 - b1
eq2 = np.log(rate2 / (1 - rate2)) - coef2 * a1 - b1
data = solve((eq1, eq2), (a1, b1))

print(f'a1 = {data[a1]:5.3f}')
print(f'b1 = {data[b1]:5.3f}')
```

```python
# f4(x)

from sympy import solve, Symbol
import numpy as np

rate1 = 0.99
rate2 = 0.01

coef1 = 3.0
coef2 = 3.4

a1 = Symbol('a1')
b1 = Symbol('b1')

eq1 = np.log(rate1 / (1 - rate1)) - coef1 * a1 - b1
eq2 = np.log(rate2 / (1 - rate2)) - coef2 * a1 - b1
data = solve((eq1, eq2), (a1, b1))

print(f'a1 = {data[a1]:5.3f}')
print(f'b1 = {data[b1]:5.3f}')
```

```python
# f3(x) + f4(x) - 1

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1.0, 5.0, 1000)
y = [x for i in x]

for i in range(len(x)):
    f3 = 1 / ( 1 + np.exp(-22.976 * x[i] + 64.332))
    f4 = 1 / ( 1 + np.exp(22.976 * x[i] - 73.522))
    y[i] = f3 + f4 - 1
    
plt.plot(x, y)
plt.xlabel('Quality')
plt.ylabel('rate')
plt.text(2.5, 0.5, 'f3(x)')
plt.text(3.2, 0.5, 'f4(x)')
plt.grid()
plt.show()
```

```python
# ( f1(x) + f2(x) - 1 ) + 0.3 ( f3(x) + f4(x) - 1 )

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1.0, 5.0, 1000)
y = [x for i in x]

for i in range(len(x)):
    f1 = 1 / ( 1 + np.exp(-0.828 * x[i] + 4.006))
    f2 = 1 / ( 1 + np.exp(6.504 * x[i] - 33.212))
    f3 = 1 / ( 1 + np.exp(-22.976 * x[i] + 64.332))
    f4 = 1 / ( 1 + np.exp(22.976 * x[i] - 73.522))
    y[i] = ( f1 + f2 - 1 ) + 0.3 * ( f3 + f4 - 1 )
    
plt.plot(x, y)
plt.xlabel('Quality')
plt.ylabel('rate')
plt.xlim(left=1.0)
plt.xlim(right=5.0)
plt.grid()
plt.show()
```
