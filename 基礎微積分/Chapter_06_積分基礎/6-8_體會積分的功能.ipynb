```python
import matplotlib.pyplot as plt
import numpy as np

# 原始函數 F(x) 的係數
a = 1

# 繪製此函數積分區間圖形
x = np.linspace(-1, 1, 1000)
y = a * x ** 2
plt.plot(x, y, color='aqua')
plt.fill_between(x, y1 = y, y2 = 0, where= (x >= 0) & ( x <= 1), facecolor='lightpink')

x_line = np.linspace(0, 0.1, 2)
y_line = x_line

# 繪製藍色水平線
for i in range(10):
    x_line[0] = i * 0.1
    x_line[1] = x_line[0] + 0.1
    y_line = [ a * (n + 0.1) ** 2 for n in x_line ]
    y_line[1] = y_line[0]
    plt.plot(x_line, y_line, color='b')
    
# 繪製藍色垂直線
for i in range(10):
    x_line[0] = i * 0.1
    x_line[1] = x_line[0]
    y_line[0] = 0
    y_line[1] = a * (x_line[0] + 0.1 ) ** 2
    plt.plot(x_line, y_line, color='b')

plt.grid()
plt.show()
```

分成 10 等分

```python
import matplotlib.pyplot as plt
import numpy as np

n = 10                             # 分成 n 等分
x = np.linspace(0, 1, n + 1)

# 較小面積 S
area = 0
for i in range(n):
    f = x[i] ** 2
    area += f
s = area / n
print(f'較小面積 = {s:6.3f}')

# 較大面積 L
area = 0
for i in range(1, n + 1):
    f = x[i] ** 2
    area += f
l = area / n
print(f'較大面積 = {l:6.3f}')

# 平均面積
average = ( s + l ) / 2
print(f'平均面積 = {average:6.3f}')
```
